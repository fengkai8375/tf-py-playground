```python
#import cv2
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt

image = misc.ascent()
# print(i)

plt.grid(False)
plt.gray();
plt.axis('off')
plt.imshow(image)
plt.show()

image2 = np.copy(image)

# print(image2)

size_x = image2.shape[0]
size_y = image2.shape[1]

print(size_x, size_y)


# CNN

filter = [ [-1, -2, -1], [0, 0, 0], [1, 2, 1]]

for x in range(1, size_x - 1):
  for y in range(1, size_y -1):
    convolution = 0
    convolution += image[x-1, y-1] * filter[0][0]
    convolution += image[x, y-1] * filter[0][1]
    convolution += image[x+1, y-1] * filter[0][2]
    convolution += image[x-1, y] * filter[1][0]
    convolution += image[x, y] * filter[1][1]
    convolution += image[x+1, y] * filter[1][2]
    convolution += image[x-1, y+1] * filter[2][0]
    convolution += image[x, y+1] * filter[2][1]
    convolution += image[x+1, y+1] * filter[2][2]
    
    if convolution < 0:
      convolution = 0
    
    if convolution > 255:
      convolution = 255
    
    image2[x, y] = convolution

plt.grid(False)
plt.gray();
plt.axis('off')
plt.imshow(image2)
plt.show()


# Maxpooling

size_x2 = size_x / 2
size_y2 = size_y / 2

image3 = np.zeros((size_x2, size_y2))

for x in range(0, size_x, 2):
  for y in range(0, size_y,  2):
    list = []
    list.append(image2[x+1, y+1])
    list.append(image2[x+1, y])
    list.append(image2[x, y+1])
    list.append(image2[x, y])
    list.sort(reverse=True)
    
    image3[int(x/2), int(y/2)] = list[0]
    
plt.grid(False)
plt.gray();
plt.axis('on')
plt.imshow(image3)
plt.show() 
```
